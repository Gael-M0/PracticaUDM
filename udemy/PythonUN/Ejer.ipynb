```python
# Tipos de datos en Python
# str = cadenas de texto
nombre = "Alan"

# int = numeros enteros
edad = 25

# float = numeros con decimales
altura = 1.75

# boolean = verdadero o falso
respira = True

# nonetype = sin valor
compania = None

print(type(nombre))
print(type(edad))
print(type(altura))
print(type(respira))
print(type(compania))
```

```python
# Ejercicio de nombres
nombre_prueba = input("Escribe tu nombre: ")
cantidad_logros = int(input("Escribe la cantidad de logros que has conseguido: "))

if cantidad_logros > 50:
    print(f'Felicidades {nombre_prueba}, eres una persona muy lograda con {cantidad_logros} logros!')
else:
    print(f'Hola {nombre_prueba}, has conseguido {cantidad_logros} logros. ¡Sigue esforzándote!')
```
