```python
'''
Ejercicio sistemas de reserva de hoteles
Hombre de cliente
Dias de estancia
tarifa diaria
indicar si el cuarto tiene vista al mar (True/False)
'''

def reserva():
    nombre = None
    Tarifa = 200
    Dias = 0
    Cuarto = False

    while True:
        print("\n===== MENÚ PRINCIPAL =====")
        print("1. Registrar huesped")
        print("2. Tipo de habitacion")
        print("3. Salir")

        opcion = input("Bienvenido elija una opcion: ")

        if opcion == "1":
            nombre = input("Ingresa tu nombre: ")
            print(f" Hola '{nombre}', registrado con éxito.")
        
        elif opcion == "2":
            if nombre:
                print(f" Hola {nombre}, bienvenido de nuevo")
                Dias = int(input("Cuantos dias planea quedarse: "))
                Cuarto = input("Desea una habitacion con vista al mar (True/False): ").strip().lower() == 'true'
                costo_total = Tarifa * Dias
                if Cuarto:
                    costo_total += 50 * Dias  # Costo adicional por vista al mar
                print(f" El costo total de su estancia es: ${costo_total}")


            else:
                print(" No hay ningún nombre registrado.")
                print("Redirigiendo a registro...")
                nombre = input("Ingresa tu nombre: ")
                print(f"Nombre '{nombre}' registrado con éxito.")
        
        elif opcion == "3":
            print("Saliendo del programa")
            break

        else:
            print("Opcion no válida. Intenta de nuevo.")


reserva()


```

```python
'''
Sistema de tienda online

Crear el detalle de un producto de una tienda online:
- Nombre del producto (str)
- Precio (float)
- Cantidad en stock (int)
- Indicar si el producto esta disponible (boolean)
imprimir los valores del producto de forma ordenada.
'''
def tienda_online():
    nombre_producto = input("Ingrese el nombre del producto: ")
    precio = float(input("Ingrese el precio del producto: "))
    cantidad_stock = int(input("Ingrese la cantidad en stock: "))
    disponible = cantidad_stock > 0

    print("\n===== DETALLE DEL PRODUCTO =====")
    print(f"Nombre del producto: {nombre_producto}")
    print(f"Precio: ${precio:.2f}")
    print(f"Cantidad en stock: {cantidad_stock}")
    print(f"Disponible: {'Sí' if disponible else 'No'}")

tienda_online()
```

```python
'''
Constantes en Python
En Python, las constantes se definen utilizando letras mayúsculas para indicar que su valor no debe cambiarse durante la ejecución del programa. Aunque Python no tiene una sintaxis específica para declarar constantes, esta convención ayuda a los desarrolladores a identificar variables que deben permanecer inalteradas.
'''

PI = 3.14159 
print(f'El valor de PI es: {PI}')

Nombre_base = "Base de datos principal"
print(f'El nombre de la base es: {Nombre_base}')

# Intentar cambiar el valor de una constante (no recomendado)
# PI = 3.14  # Esto no es recomendable ya que PI es una constante
Pi = 3.14  # Esto no es recomendable ya que PI es una constante
print(f'El nuevo valor de PI es: {Pi}')
```
