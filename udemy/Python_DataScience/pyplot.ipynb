```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10,0.1)
y=np.sin(x)
z=np.cos(x)
plt.grid()
plt.title('Mi primera gr√°fica')
plt.plot(x,y,label='sin(x)')
plt.plot(x,z,label='cos(x)')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.xlim(0,4)
plt.legend()
plt.show()
```
