```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10, 0.1)
y = np.sin(x) #Seno de x
z = np.cos(x) #Coseno de x
plt.grid() #Añade una cuadrícula al gráfico
plt.title('Seno y Coseno') #Título del gráfico
plt.plot(x, y, label='Seno') #Etiqueta la línea del seno
plt.plot(x, z, label='Coseno') #Etiqueta la línea del coseno
plt.xlabel('Eje X') #Etiqueta del eje X
plt.ylabel('Eje Y') #Etiqueta del eje Y
plt.xlim(0, 4) #Límites del eje X
plt.legend() #Muestra la leyenda
plt.show() #Muestra el gráfico, sin datos extraños
```

## Otros tipos de graficos

### Scatter plot

```python
x = np.random.random([50])
y = np.random.random([50])
plt.scatter(x, y)
plt.show()
```

### Bar plot

```python
opciones = ['si', 'no']
resultados = [40,50]
plt.bar(opciones, resultados)
plt.show()
```

## Histogramas

```python
x = np. random.random([50]) #50 números aleatorios entre 0 y 1
plt.hist(x, 5) #5 barras
plt.show()
```

### Blox plot

```python
plt.boxplot(x)
plt.show()
```

### Pie plot


```python
plt.pie(resultados, labels=opciones)
plt.show()
```

## Tamaños, colores, estilos de linea y marcadores

```python
y = np.array([1, 4, 9, 16, 25])
x = np.array([1, 2, 3, 4, 5])
```

```python
plt.figure(figsize=(12,6)) #Tamaño de la figura
plt.grid() 
plt.plot(x, y,'o:g',c='r',mfc='b',mec='g',ms=20) #Marca con un círculo cada punto | Línea de puntos | Color verde 
# C es el color de la línea | mfc es el color de relleno de la marca | mec es el color del borde de la marca | ms es el tamaño de la marca
plt.show()
```
